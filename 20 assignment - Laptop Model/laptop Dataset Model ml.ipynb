```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('laptop_price - dataset.csv')

print(" First 10 rows of the dataset:")
print(df.head(10))

print("\n Shape of the dataset:")
print(df.shape)
```

```python
# Independent Features
X = df.drop("Price (Euro)", axis=1)

# Dependent Feature
y = df["Price (Euro)"]

print("Independent Features:")
print(X.columns)
print("Shape of the independent features:")
print(X.shape)

print("\nDependent Feature:")
print(y.name)
print("Shape of the dependent feature:")
print(y.shape)
```

```python
# One Hot Encoding of input categorical feature 
x_cat_col = df.select_dtypes(include=['str'])
x_input_encoded = pd.get_dummies(x_cat_col)

# Encoded input data to integer type
x_ip_encoded_integer = x_input_encoded.astype(int)

# 5 rows of encoded integer
print("First 5 Rows of Encoded Independent Feature Converted in Integer:")
print(x_ip_encoded_integer.head(5))
```

```python
from sklearn.preprocessing import StandardScaler

# Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x_ip_encoded_integer)

# Convert scaled data into DataFrame
scaled_df = pd.DataFrame(X_scaled)

# Display first 5 rows
print("Display First 5 Rows of Scaled Dataset:")
print(scaled_df.head())
```

```python
# Import library
from sklearn.model_selection import train_test_split

# Input features (X) and Target variable (y)
X = df.drop("Price (Euro)", axis=1)
y = df["Price (Euro)"]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    scaled_df,
    y,
    test_size=0.33,
    random_state=42
)

# Print the shapes
print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```

```python
# Import Linear Regression
from sklearn.linear_model import LinearRegression

# Create the model
model = LinearRegression()

# Train (fit) the model
model.fit(X_train, y_train)
# Print the intercept
print("Intercept:")
print(model.intercept_)

# Print the coefficients
print("\nCoefficients:")
for feature, coef in zip(X_train.columns, model.coef_):
    print(f"{feature}: {coef}")
```

```python
# Make predictions on the test data
y_pred = model.predict(X_test)

# Print first 10 predicted values
print("First 10 Predicted Values:")
print(y_pred[:10])

# Print first 10 actual values
print("\nFirst 10 Actual Values:")
print(y_test.iloc[:10].values)

# Compare predicted and actual values
comparison = pd.DataFrame({
    "Actual Price": y_test.iloc[:10].values,
    "Predicted Price": y_pred[:10]
})

print("\nComparison of Actual vs Predicted Values:")
print(comparison)
```

```python
# Import r2_score
from sklearn.metrics import r2_score

# Calculate R² score
r2 = r2_score(y_test, y_pred)

# Print R² score
print("R² Score:", r2)
```

```python
# Import joblib
import joblib

# Save the trained Linear Regression model
joblib.dump(model, "LR_Laptop_price.pkl")

# Save the StandardScaler
joblib.dump(scaler, "scaler_laptop.pkl")

# Save the list of feature columns
joblib.dump(X_train.columns.tolist(), "columns_laptop.pkl")

print("Model and preprocessing objects saved successfully!")
```
