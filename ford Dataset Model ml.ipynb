```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import joblib

# Load dataset
df = pd.read_csv("ford_car_dataset - ford_car_dataset.csv")

# Input and Output
X = df.drop("price", axis=1)
y = df["price"]

# One-Hot Encode
X = pd.get_dummies(X)

# Save feature names
encoded_columns = X.columns.tolist()

# Scale only numeric columns
numeric_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

scaler = StandardScaler()
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.33,
    random_state=42
)

# Train Model
model = LinearRegression()
model.fit(X_train, y_train)

# Save Files
joblib.dump(model, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(encoded_columns, "columns.pkl")

print("Saved Successfully")
```
